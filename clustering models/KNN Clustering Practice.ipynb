### Prepare the data for modelling

-For culstering we will us RFM which is said to commonly used notion for retailer data.

-Recency,Frequency,Monetary.

-Using RFM we will extract features which can be helpful for clustering

```python
data['amount'] = data['UnitPrice']*data['Quantity']
```

```python
data.head()
```

```python
Monetary= data.groupby('CustomerID')["amount"].sum()
Monetary= Monetary.reset_index()
Monetary
```

```python
frequency = data.groupby('CustomerID')["InvoiceNo"].count()
frequency = frequency.reset_index()
frequency.columns= ['CustomerID','frequency']
frequency.head()
```

```python
df = pd.merge(Monetary,frequency , on='CustomerID', how='inner' )
```

```python
df.head()
```

```python
data.info()
```

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'],format="%d-%m-%Y %H:%M")
```

```python
max(data['InvoiceDate'])
```

```python
data['date_diff'] = max(data['InvoiceDate'])-data['InvoiceDate']
```

```python
data.head()
```

```python
recency = data.groupby('CustomerID')['date_diff'].min()
recency = recency.reset_index()
recency.columns = ['CustomerID','recency']
recency.head()
```

```python
retail_df= pd.merge(df,recency , on='CustomerID',how='inner' )
```

```python
retail_df.head()
```

```python
retail_df['recency'] = retail_df['recency'].dt.days
```
